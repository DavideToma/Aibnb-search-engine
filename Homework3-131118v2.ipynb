```python
import pandas as pd
import unicodecsv as csv # We also import the unicodecsv library to handle any Unicode objects in the CSV report...as csv.
import nltk
import string

from nltk import word_tokenize
from nltk.corpus import stopwords

nltk.download('stopwords')
```

STEP 1 - DATA

```python
DIR = 'data/docs/' # C:/Users/david/Desktop/DATA SCIENCE/AMD/documents/

df = pd.read_csv("data/Airbnb_Texas_Rentals.csv", sep=",") # we read the dataset
```

```python
df = df.drop('Unnamed: 0',axis = 1) #this column is useless for our analisys
df = df[pd.notnull(df['description'])] #If the row has no description, for us is useless, so we drop it
df = df.replace(r'\\n',' ', regex=True) #Remove all the new lines in the dataframe
df['language'] = 'nd'
df.head()
```

STEP 2 - WRITE DOCUMENTS

```python
for row in df.iterrows():
    file_name = DIR + 'doc_'+str(row[0])+'.tsv'
    with open(file_name,'wb') as file: # the wb means we are writing a file in binary mode.
        csv_writer = csv.writer(file, delimiter = '\t') #now we are writing the files as tab separated files.
        csv_writer.writerow(list(row[1].values)) #writerow want a list object ad input, so we used list.
        file.close() #Once the file is written, we close it, so we free some memory
```

```python
stopEn = stopwords.words('english') #import english stopwords
stopSp = stopwords.words('spanish') #import english stopwords

stopEn.extend(["another"]) #Add other words that are not included in the stopwords list but that we want to exclude

stemEn = nltk.stem.SnowballStemmer('english')#Defining an english stemmer
stemSp = nltk.stem.SnowballStemmer('spanish')#Defining a spanish stemmer
docdict = {}

for i, row in df[:10000].iterrows(): #We limit to 5 just to work quickly

    #Remove punctuation
    text = row['description'].translate({key: None for key in string.punctuation})
    
    if( (len(set(text.lower().split(' ')) & set(stopSp)) > 5) & (len(set(text.lower().split(' ')) & set(stopEn)) < 5)):
        #The text here is Spanish, we can decide what to do
        df.loc[i,'language'] = 'sp'
        
        #Remove all the stopwords and all the non alphabetic words
        tostem = ([x.lower() for x in text.split(' ') if x.lower() not in stopSp and x.isalpha()])
        
        #Stem all the words
        docdict[i] = [stemSp.stem(word) for word in tostem]
    
    else:
        #The text here is Englush, we can decide what to do
        df.loc[i,'language'] = 'en'
        
        #Remove all the stopwords and all the non alphabetic words
        tostem = ([x.lower() for x in text.split(' ') if x.lower() not in stopEn and x.isalpha()])
        
        #Stem all the words
        docdict[i] = [stemEn.stem(word) for word in tostem]
    


# docdict
```
